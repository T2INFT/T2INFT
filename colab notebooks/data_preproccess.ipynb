```python
Image.MAX_IMAGE_PIXELS = None
```

```python
img_path = '/content/dataset/image/*'

for file_name in glob.glob(img_path):
  img = Image.open(file_name)
  if img.mode == 'RGB':
    img = img.resize((512, 512))
    img.save('{}{}'.format('/content/processed_data/', file_name.replace('/content/dataset/image/', '')))
```
